```python
import pandas as pd
import folium
```

фолиум эт крч питоновская приколюмба для визуализации гео-пространственных данных 
п-ссс она создана по мотивам карт таро, Datawrangling а сама карта с OpenStreetMap (https://www.openstreetmap.org)

```python
world_map = folium.Map()
world_map
```

```python
# теперь откроем россиюшку 
latitude = 64.6863136
longitude = 97.7453061

rossiyushka = folium.Map(location = [latitude, longitude], zoom_start = 4)
# location - указываем широту и долготу 
# zoom_start - просто стартовый зум 
# карту можно и нужно трогать типа колесиком мышки увеличивать уменьшать и двигать как тебе угодно 
# вообще тут много всяких параметров (особенно почекай параметр "tiles" это стили карт)    полезно почитать (https://python-visualization.github.io/folium/latest/reference.html)

rossiyushka
```

```python
# вот примерчик с использованим tiles 
rossiyushka = folium.Map(location = [latitude, longitude], zoom_start = 4, tiles='cartodb positron')

rossiyushka
```

```python
# терь попробуем найти славный колледж и поставить на него маркер 
latitude_nkeivt = 55.068937
longitude_nkeivt = 82.908596
nkeivt = folium.Map(location = [latitude_nkeivt, longitude_nkeivt], zoom_start = 20)

folium.Marker(location=[latitude_nkeivt,longitude_nkeivt], popup = "НКЭиВТ", icon=folium.Icon(color = 'red')).add_to(nkeivt)

nkeivt
```
